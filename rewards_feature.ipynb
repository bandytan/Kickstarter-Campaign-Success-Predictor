```python
import pandas as pd 
import ast
import re
import numpy as np
df = pd.read_csv("./data/Kickstarter_merged.csv")
df.head(3)
df_copy = df.copy()
```

```python
df_copy["reward_tiers"] = df_copy["rewards"].apply(lambda x : len(ast.literal_eval(x)))
df_copy.head()
```

```python
# Rearange reward_tiers column to beside rewards
cols = df_copy.columns.tolist()
cols = cols[:6] + [cols[-1]] + cols[6:-1]
df_copy = df_copy[cols]
df_copy
```

```python
df_copy["all_reward_amount"] = np.empty((len(df), 0)).tolist()
# df_copy["all_reward_amount"][0]
```

Generate all reward amounts for each listing

```python
# for i in range(len(df_copy)):
# money = re.compile('|'.join([
#   r'^\$?(\d+, \d{1,3})$',  # e.g., $.50, .50, $1.50, $.5, .5
#   r'^\$?(\d+)$',           # e.g., $500, $5, 500, 5
# ]))

for i in range(len(df_copy)):
    all_reward_amount = []
    dict_list = ast.literal_eval(df_copy.iloc[i, 5])

    for dict in dict_list:
        values_string = str(dict.values())
        # reward_title = re.search(r"Pledge S\$ \d{1,3}?(, \d{1,3}) or more", values_string)
        reward_title = re.search(r"Pledge S\$ \d{1,3}(,\d{1,3})? or more", values_string)

        if reward_title is not None:
            # print(reward_title.group())
            reward_amount = re.search(r"\d{1,3}(,\d{1,3})?", reward_title.group())
            if reward_amount is not None:
                all_reward_amount.append(reward_amount.group())
        else:
            # print("reward title is none")
            all_reward_amount.append(0)
    df_copy["all_reward_amount"][i] = all_reward_amount
    
```
